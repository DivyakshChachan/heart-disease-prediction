# 01 - Data Exploration
Load and inspect the UCI/Kaggle heart disease dataset.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
data_path = os.path.join('..','data','heart_disease_dataset.csv')
df = pd.read_csv(data_path)
df.head()
```

```python
df.info()
```

```python
df.describe().T
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm', center=0, annot=False)
plt.title('Correlation Heatmap')
plt.show()
```
